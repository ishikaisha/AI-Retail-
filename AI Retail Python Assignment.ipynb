## Data Loading and Preprocessing





The data set contains information that must be preprocessed in this sense and adapted for analysis for various reasons. The following steps have been undertaken to prepare the data:

1. Preparing and Conversion of considered columns
   
* The dataset also has multiple categorical values, which have to be transformed into numerical values for the analytical process. Concerning these columns, one hot encoding was used to transform them into binary variables. For instance:

* Some categorical variables such as Gender, Age, Payment_Method_Credit/Debit were converted into exclusive binary variables (Gender_Male, Age_Gen X, Payment_Method_Credit/Debit.YES). This helps to fit well in the statistical and machine learning operations that take results in numbers.
  
Also, features that were originally categorical, for example, Annual_Salary where ‘Low’, ‘Medium’, and ‘High’ were the categories used, were encoded to new binary features each representing a category.

2. Handling Redundant Columns
After encoding, the one-hot encoded derived columns with similar information were deleted from the dataset by recognizing some of them as the original categorical columns.

* This study deleted the Gender variable that was originally used while making a binary variable like Gender_Male.
* In the same way, nominal features such as Age and Annual_Salary transformed into their respective dummy binary features.
  
4. Data Standardization
   
However, no other data preprocessing steps are possible since the dataset does not contain any continuous numerical variables that can be called for standardization (such as z-scoring and min-max scaling ). But if add new quantitative features, can be obtained with numerical transformations, these quantitative features can be standardized to ensure that they are within the same range as the other independent variables.

5. Eliminate Duplicate Capabilities
   
* These were retained as separate features to enable an examination of the effects of each specific AI tool including; AI_Tools_Used_Chatbots, AI_Tools_Used_Virtual_Assistant, and AI_Tools_Used_Voice&Photo_Search.

* As will be seen, no features were specifically determined for merging with the current data structure identified at the time of writing this paper. However, features are only allowed on the aggregated level if more general insights are needed (for example: a summative feature AI_Tools_Used_Total, which sums up the total usage of all AI tools).

6. Final Dataset
After data cleaning, the current dataset is composed of 32 columns, and each of them reflects a clear and valuable feature (Jain et al. 2021). Some original categorical columns are redundant and then removed, consequently, the dataset has been saved in the name Processed_AI_in_Retail_Dataset.csv. This processed dataset therefore qualifies for exploratory data analysis as well as statistical or even machine learning computations.




### Code Expalanation 






**Load the Dataset**:

* The data set is imported with pandas by using the pandas.read_csv by specifying latin1 to ensure the handling of any special characters.

**Explore the Dataset**:

* Shows simple characteristics of the tables (number of columns, types of columns, non-null values count).
* Generates some sample rows for reference at a glance.

**Handle Missing Values**:

* Remedial action on missing value(s) is done by identifying them on the application of the missing value imputation technique.
* Replacing missing data with mean for numeric variables and mode for the categorical variables.

**Convert Data Types**:

* Converts integer encoded features stored as strings to their respective numeric data type.

**Encode Categorical Variables**:

* Executes one-hot encoding for a categorical variable using the pd.get_dummies.
  
**Save the Processed Dataset**:

The last step is, the cleaned and processed dataset will be saved in a new CSV file named Processed_AI_in_Retail_Dataset.csv.

                                                               

```python


# Importing necessary libraries
import pandas as pd
import numpy as np

# Load the dataset
file_path = 'AI_in_Retail_Dataset.csv'  
try:
    # Loading the dataset with appropriate encoding
    data = pd.read_csv(file_path, encoding='latin1')
    print("Dataset successfully loaded!")
except Exception as e:
    print(f"Error loading dataset: {e}")

# Displaying basic information about the dataset
print("\nBasic Information:")
data.info()

# Checking the first few rows of the dataset
print("\nSample Rows:")
print(data.head())

# Checking for missing values
print("\nMissing Values:")
print(data.isnull().sum())

# Handling missing values (if any)
# If no missing values, this step can be skipped
# Example: Replace missing values with the column mean for numeric data
for column in data.columns:
    if data[column].isnull().sum() > 0:
        if data[column].dtype in ['float64', 'int64']:
            data[column].fillna(data[column].mean(), inplace=True)
        else:
            data[column].fillna(data[column].mode()[0], inplace=True)

# Confirming missing values have been handled
print("\nMissing Values After Handling:")
print(data.isnull().sum())

# Converting data types if needed
# Example: Convert numerical data stored as objects to appropriate numeric types
for column in data.select_dtypes(include=['object']).columns:
    if data[column].str.isnumeric().all():
        data[column] = pd.to_numeric(data[column])

# Encoding categorical variables (if needed)
# Example: Encoding binary categorical variables as 0 and 1
categorical_columns = data.select_dtypes(include=['object']).columns
data_encoded = pd.get_dummies(data, columns=categorical_columns, drop_first=True)

# Displaying the processed dataset
print("\nProcessed Dataset (First Few Rows):")
print(data_encoded.head())

# Saving the processed dataset to a new CSV file for future use
processed_file_path = 'Processed_AI_in_Retail_Dataset.csv'
data_encoded.to_csv(processed_file_path, index=False)
print(f"\nProcessed dataset saved to {processed_file_path}.")


```

```python



# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report

# Load the processed dataset
file_path = 'Processed_AI_in_Retail_Dataset.csv'
data = pd.read_csv(file_path)

# EDA Section
print("Exploratory Data Analysis (EDA)")

# Descriptive Statistics
print("\nDescriptive Statistics:")
print(data.describe())

# Correlation Matrix
print("\nCorrelation Matrix:")
correlation_matrix = data.corr()
print(correlation_matrix)

# Visualizing Correlation Matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()

# Frequency Distribution of AI Tools Used
ai_tools_columns = [
    "AI_Tools_Used _Chatbots_YES",
    "AI_Tools_Used_Virtual_Assistant_YES",
    "AI_Tools_Used_Voice&Photo_Search_YES"
]
ai_tools_usage = data[ai_tools_columns].sum()

plt.figure(figsize=(8, 5))
ai_tools_usage.plot(kind="bar", color=['skyblue', 'lightgreen', 'salmon'])
plt.title("Frequency of AI Tools Used")
plt.xlabel("AI Tools")
plt.ylabel("Number of Consumers")
plt.xticks(rotation=0)
plt.show()

# Distribution of Payment Methods
payment_methods = [
    "Payment_Method_COD_YES",
    "Payment_Method_Ewallet_YES",
    "Payment_Method_Credit/Debit_YES"
]
payment_usage = data[payment_methods].sum()

plt.figure(figsize=(8, 5))
payment_usage.plot(kind="bar", color=['gold', 'blue', 'green'])
plt.title("Distribution of Payment Methods")
plt.xlabel("Payment Methods")
plt.ylabel("Number of Consumers")
plt.xticks(rotation=0)
plt.show()

# Analyzing Product Preferences vs AI Tools Used
product_columns = [
    "Product_Category_Appliances_YES",
    "Product_Category_Electronics_YES",
    "Product_Category_Groceries_YES",
    "Product_Category_Personal_Care_YES",
    "Product_Category_Clothing_YES"
]
for tool in ai_tools_columns:
    plt.figure(figsize=(10, 6))
    data.groupby(tool)[product_columns].sum().plot(kind='bar', stacked=True)
    plt.title(f"Product Preferences by {tool}")
    plt.ylabel("Count")
    plt.xlabel(tool)
    plt.xticks(rotation=0)
    plt.legend(title="Product Categories")
    plt.show()

# Advanced Analysis Section
print("Advanced Analysis")

# Clustering: KMeans Clustering on AI Usage
print("\nKMeans Clustering on AI Usage")

# Selecting relevant columns for clustering
clustering_data = data[ai_tools_columns]

# Standardizing the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(clustering_data)

# Applying KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(scaled_data)
data['Cluster'] = kmeans.labels_

# Visualizing Clusters
plt.figure(figsize=(8, 5))
sns.scatterplot(
    x=scaled_data[:, 0], 
    y=scaled_data[:, 1], 
    hue=data['Cluster'], 
    palette="deep"
)
plt.title("Clusters of AI Tool Users")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend(title="Cluster")
plt.show()


# Exploring Demographics
age_columns = ["Age_Gen X", "Age_Gen Z", "Age_Millennials"]
age_distribution = data[age_columns].sum()

plt.figure(figsize=(10, 6))
age_distribution.plot(kind="bar", color=['purple', 'orange', 'pink'])
plt.title("Age Group Distribution")
plt.xlabel("Age Groups")
plt.ylabel("Number of Consumers")
plt.xticks(rotation=0)
plt.show()


# Bar Chart: Distribution of Product Preferences
product_columns = [
    "Product_Category_Appliances_YES",
    "Product_Category_Electronics_YES",
    "Product_Category_Groceries_YES",
    "Product_Category_Personal_Care_YES",
    "Product_Category_Clothing_YES"
]
product_preferences = data[product_columns].sum()

plt.figure(figsize=(10, 6))
product_preferences.plot(kind="bar", color=['purple', 'orange', 'pink', 'lightblue', 'green'])
plt.title("Distribution of Product Preferences")
plt.xlabel("Product Categories")
plt.ylabel("Number of Consumers")
plt.xticks(rotation=45)
plt.show()




```
